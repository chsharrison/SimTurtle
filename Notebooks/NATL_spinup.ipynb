```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc
import os
```

```python
# make default figure size 2x bigger, can toggle this
import matplotlib
matplotlib.rcParams['savefig.dpi'] = 2 * matplotlib.rcParams['savefig.dpi']
```

```python
# visualize this area

path = "/glade/scratch/chsharri/Hi-res/For_CMS/"
date = "1-07-04"
myfile = 'g.e11.G.T62_t12.eco.006.pop.h.000' + date + '.regrid_surf.nc' 

mydata = nc.Dataset(path+myfile,'r')
lat = mydata.variables['lat'][:]
lon = mydata.variables['lon'][:]
u = mydata.variables['UVEL'][:]
v = mydata.variables['VVEL'][:]
sst = mydata.variables['SST'][:]

# convert to m/s
u = u/100.0
v = v/100.0
```

```python
skip = 5
lat_min = 23
lat_max = 38
lon_min = 275 - 360
lon_max = 320 - 360
```

```python
# get North Atlantic subset
print lon_min, lon_max, lat_min, lat_max
ilon1= np.argmax(lon>lon_min)
ilon2 = np.argmax(lon>lon_max)
ilat1 = np.argmax(lat>lat_min)
ilat2 = np.argmax(lat>lat_max)
print ilon1,ilon2,ilat1,ilat2
print lon[ilon1], lon[ilon2],lat[ilat1], lat[ilat2]
```

```python
plt.pcolormesh(lon[ilon1:ilon2], lat[ilat1:ilat2], sst[ilat1:ilat2,ilon1:ilon2], vmin = 20, vmax = 30)
plt.axis([lon_min, lon_max, lat_min, lat_max])
cb1 = plt.colorbar()
#cb1 = plt.colorbar(orientation='horizontal')
Q = plt.quiver(lon[ilon1:ilon2:skip],lat[ilat1:ilat2:skip],u[ilat1:ilat2:skip,ilon1:ilon2:skip],
           v[ilat1:ilat2:skip,ilon1:ilon2:skip], scale = 15)

# need to work on getting this key to work:
#qk = plt.quiverkey(Q, 0.5, 0.92, 2, r'$2 \frac{m}{s}$', labelpos='W',
#                   fontproperties={'weight': 'bold'})

plt.axes().set_aspect('equal')
plt.title(date)
plt.xlabel('Lon')
plt.ylabel('Lat')
```

## Check regridding against original

```python
# check against original

# get original velocity field

path = '/glade/scratch/mclong/hi-res-eco/g.e11.G.T62_t12.eco.006/ocn/hist/'
myfile = 'g.e11.G.T62_t12.eco.006.pop.h.000' + date + '.nc' 
mydata = nc.Dataset(path+myfile,'r')
ulat = mydata.variables['ULAT'][:]
ulon = mydata.variables['ULONG'][:]
sst1 = mydata.variables['TEMP'][0,0,:,:]
tlat = mydata.variables['TLAT'][:]
tlon = mydata.variables['TLONG'][:]
```

```python
uvel = mydata.variables['UVEL'][0,0,:,:]
vvel = mydata.variables['VVEL'][0,0,:,:]

# convert u and v to m/s
uvel = uvel/100.0
vvel = vvel/100.0
```

```python
# need to make tlon have same range as ulon
tlon[tlon>180] = tlon[tlon>180] -360
```

```python
lulon = ulon[1,:]
lulat = ulat[:,1]
print lulon, lulat
```

```python
# get North Atlantic subset for native POP grid
print lon_min, lon_max, lat_min, lat_max
iulon1= np.argmax(lulon>lon_min)
iulon2 = np.argmax(lulon>lon_max)
iulat1 = np.argmax(lulat>lat_min)
iulat2 = np.argmax(lulat>lat_max)
print iulon1,iulon2,iulat1,iulat2
print lulon[iulon1], lulon[iulon2],lulat[iulat1], lulat[iulat2]
```

```python
plt.pcolormesh(tlon[iulat1:iulat2,iulon1:iulon2], tlat[iulat1:iulat2,iulon1:iulon2], sst1[iulat1:iulat2,iulon1:iulon2],
               vmin = 20, vmax = 30)
# this has to be before the quiver call to work:
plt.colorbar()
plt.quiver(ulon[iulat1:iulat2:skip,iulon1:iulon2:skip],ulat[iulat1:iulat2:skip,iulon1:iulon2:skip],
           uvel[iulat1:iulat2:skip,iulon1:iulon2:skip],vvel[iulat1:iulat2:skip,iulon1:iulon2:skip],scale = 15)
plt.axis([lon_min, lon_max, lat_min, lat_max])

plt.axes().set_aspect('equal')
plt.title(date)
plt.xlabel('Lon')
plt.ylabel('Lat')
```
